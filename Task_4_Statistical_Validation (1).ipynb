# Hypothesis Testing: Impact of Toss on Match Outcome

```python
matches['toss_match_win'] = matches['toss_winner'] == matches['winner']

matches['toss_match_win'].value_counts()
```

## Create Contingency Table

The contingency table summarizes the relationship between toss decisions and match outcomes.

```python
contingency_table = pd.crosstab(
    matches['toss_decision'],
    matches['toss_match_win']
)

contingency_table
```

## Perform Chi-Square Test of Independence

The Chi-Square Test is used to determine whether there is a statistically significant relationship between toss decisions and match outcomes.

```python
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square Statistic:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
```

## Decision Rule

Significance Level (α) = 0.05

- If p-value < 0.05 → Reject H₀
- If p-value ≥ 0.05 → Fail to Reject H₀

```python
alpha = 0.05

if p < alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

## Business Interpretation

The Chi-Square Test results indicate that the p-value is less than the significance level (α = 0.05). Therefore, the Null Hypothesis is rejected.

This suggests that there is a statistically significant association between toss decisions and match outcomes in the IPL dataset.

From a business perspective, toss-related decisions may influence match results. However, toss outcomes should not be considered the sole factor affecting success, as team strength, player performance, venue conditions, and match strategies also play important roles.

These findings support the use of data-driven decision-making when evaluating match strategies and team performance.
